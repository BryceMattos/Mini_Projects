```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
disney = yf.Ticker("DIS")
```

```python
disney_history = disney.history(period = "max")
disney_price_history = disney_history.iloc[:, :4]
```

```python
disney_price_history.reset_index(inplace=True)
x = disney_price_history['Date']
y = disney_price_history['Open']

plt.plot(x, y)
plt.title('Disney Stock Price Overtime', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Stock Price', fontsize = 12)
```
